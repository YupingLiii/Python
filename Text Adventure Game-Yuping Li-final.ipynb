```python
import random

#Defined the user_name and gave the introduction of the adventure.
user_name = input(prompt = """Before we begin the game, could you tell me what's your name?
Your name: """)
user_name = user_name.capitalize()

#Defined the game background.
#The idea of game come from a Chinese novel. https://www.wuxiaworld.com/novel/terror-infinity.
print(f"""

Hello, {user_name}, welcome to the world of Terror Infinity, a game let your experience different horror movies! 

Every time you finish a mission–that is you lived through a movie–you will earn 1000 points. 
You can exchange these points for everything.

""")

#Defined variables that related to tasks in the all stages.
remaining_time = 120
survive_time = 0
mission_target = "Captain Matthew"
points  = 0

#Defined the start stage in Resident Evil 1. This stage will continue until the Red Queen is closed.
def start_resident_evil_1():
    """This is the stage that you go before shut down the queen red."""
    
    #Introduction about gamer's mission and background in resident evil.
    print(f"""
{'*' * 50}
    
Hello,{user_name}, you're a gamer in the resident evil I world now.
In this world, You can't leave stay further than 300 feet from {mission_target}.
Try to survive in this world for {remaining_time} minutes.
    
{'*' * 50}
    
You' re on the train to the entrance of the Hive, which is the lab belong to the Umbrella Corporation. 
After enter in the Hive, Captain Matthew and a few mercenaries got the elevator door open. But it was pitch black inside. 
They threw down a flare and saw that the elevator was at the bottom with the wires cut off. 
"Looks like we're taking the stairs."
    
{'*' * 50}
    """)
    
    #Make the first choiec for gamer in Stage 1. Choose to follow the Captain or not.
    choice_1_1 = input (prompt = """
Do you want to follow the Captain? <Y/N> 
Your choice: """)

    #convert choice to lower letters.
    choice_1_1 = choice_1_1.casefold()
    
    #According to the common rule, if the gamer does not stay around the mission target, the gamer will be failed.
    if 'y' in choice_1_1:
        print("""
You followed the team. After went through a few doors you arrived at a control room.
After preparation, Captain Matthew let out a deep breath and waved to the people outside, "Move up."
""")
        
        #In the movie, the Captain will die in the laser corridor, the player could choose whether to change the plot.
        choice_1_2 = input (prompt = """
Do you want to alert the Captain about the potential dangers in the corridor? <Y/N>
Your choice: """)
        
        #convert choice to lower letters.
        choice_1_2 = choice_1_2.casefold()
        
        #If the gamer alert the captain about the  potential dangers, it will change the plot, and the Captain may survive.
        #If the gamer does not do anything, everything will not change.
        if 'y' in choice_1_2:
            stage_laser_room()

        #When players choose not to interfere with the plot, the plot will be the same as the initial movie's direction.     
        elif 'n' in choice_1_2:
            print("\nEverything happened like the plot in the movie. No one survived in the laser corridor.")
            stage_no_captain()
        else:
            print("\nPlease input 'Y' or 'N'. You can play again.")
            start_resident_evil_1()
    
    #fail condition for first choice in stage 1.    
    elif 'n' in choice_1_1:
        print(f"""\nYou are too far from the {mission_target}.""")
        fail()
    
    else:
        print("\nPlease put 'Y' or 'N'. You can play again.")
        start_resident_evil_1()

#Defined stage that all the plots happened in the laser corridor.
#The player who choose to alert the captain needs to go in the laser corridor. The risk of changing plot.
#Since the player choose to save other's life, so if the player put invaild input, the player can get chance play again from the start of this stage.
def stage_laser_room():
#Long line that indicates the start of a new scene.
    print(f"""
{'*' * 100}

Captain came out from the corridor and silently stared at {user_name}.
"Okay. Now you're trying to stop us from rebooting the Red Queen? You, come with us." He pointed to you.
Once everyone was inside,the doors to both sides of the corridor closed up.  
The glass walls dimmed, then a thin laser appeared between the walls and started moving toward you.
""")
    
    #The first laser will be at the neck height.
    choice_laser_1 = input(prompt = f"""
Which move you want to take to dodge the first laser?
(Please input the number or name of your choice.)
1. Lie down on the ground
2. Crouch down
3. Stay there without moving
Your choice: """)
    
    #Convert choice to lower letters.
    choice_laser_1 = choice_laser_1.casefold()
    
    #Conditional statement about the first choice in laser room stage.
    if choice_laser_1 == '1' or choice_laser_1 == '2' or  'down' in choice_laser_1 or 'lie' in choice_laser_1 or 'crouch' in choice_laser_1:
        print(f"""
The laser barely touched your shoulder. You could feel that sharpness and heat passing through your shoulder. 
                
{'*' * 50}
                
By remebering the movie plot, you survived from the second laser. 
Then the third laser was comming. Only you and Captain Matthew are still alive.
""")
                
        #The second choice in Laser stage. Give multiple choice questions to decide if player can change the plot.
        choice_laser_2 = input(prompt = f"""
Which move you want to take to survive?
(Please input the number or name of your choice.)
1. Stay near the door and pray.
2. Grabbed captain and ran near the door.
3. Trying to dodge the laser as before.
Your choice: """)
        choice_laser_2 = choice_laser_2.casefold()
        
        #Conditional statement about the choice of third laser.
        #The third laser will come as a net shape. So only staying near the door can survive.  
        #But the player can choose whether save the Captain's life.
        if choice_laser_2 == '1' or 'stay near the door' in choice_laser_2:
            print("\nYou survived but Captain died.Therefore, other people became a little upset with you")
            stage_no_captain()
                    
        elif choice_laser_2 == '2' or 'captain' in choice_laser_2:
            stage_captain_alive()
                    
        elif choice_laser_2 == '3' or 'dodge' in choice_laser_2:
            print("\nThe laser turned into the shape of a net and killed you.")
            fail()
        else:
            print("\nPlease follow the input instruction!!!")
            stage_laser_room()
    
    #Player that did not react to the first laser will die.   
    elif choice_laser_1 == '3' or 'stay' in choice_laser_1:
        print("\nThe laser went toward you and kill you.")
        fail()
    else:
        print("\nPlease follow the input instruction!!!")
        stage_laser_room()
    
#Defined the Stage 3. In this stage, Captain already died. And player have to follow new mission target.
#In this stage, player needs to face a more dangerous environment.
def stage_no_captain():

    #Long line that indicates the start of a new scene.
    #Warning about error input
    print(f"""
{'*' * 100}
Attention: If you do not follow the input instruction in this stage, it may cause you to fail.
""")
    
    #Defined the variables in new stage.
    mission_target = "Alice"
    survive_time = 50
    left_time = remaining_time - survive_time

    #Stage introduction
    print(f"""
Since the Captain Matthew died, you need to follow with the new target: {mission_target}.
You still need to survive in this movie for {left_time} minutes.
    
{'*' * 50}
    
After shut down the Red Queen, the team prepared to leave the Hive.
Consider the new mission target is {mission_target}, you decided to follow the team.
""")
    
    #Yes/No Question that asks if the player wants to help the NPC. 
    choice_2_1 = input(prompt = f"""
During the fierce battle that took place in the storage area, the team intended to open the cargo elevator to escape.
But the tech member was too nervous to input the right code. 
Do you want to help input the code? <Y/N>
Your Choice: """)
    
    #Convert choice_2_1 to lower letters.
    choice_2_1 = choice_2_1.casefold()
    
    #Since the Captain died, if player does not help input the code, player will fail.
    #However, although the player helps input the code, player can not enter the right password. 
    #In the movie, the person who opened the elevator door died. 
    if 'y' in choice_2_1:
        print("""
"What's the code?". You yelled to the tech member.
"94..941..11..20." He anserwed.
""")
        password_chances = 1
        
        #Due Because of time constraints, players only have two chances to enter. 
        #Wrong input will cause the NPC will be anxious to interrupt the player.
        #Using while loop to achieve chance reduction.
        while password_chances >= 0:
            
            password = input(prompt = f"""
Please enter your password(6 digits number).
Password: """)
            
            #Make sure the input of password will be a number input.
            try:
                password = int(password)

            except:
                print("\nYou touched the wrong button causing the alarm to go off and many walkers pounced on you when they heard the sound. You die.")
                fail()
                break
            
            #Conditional statement that used to determine if the password was entered correctly.
            #Two chances for entering the password.
            #Small eggs, the password is my boyfriend's birthday. Haha.
            if password == 941120:
                print("\nYou open the door of cargo elevator. But it was fulled with zombies.")
                fail()
                break
                
            elif password != 941120 and password_chances > 0:
                print(f"""
Wrong password.
You have {password_chances} attempt(s) remaining.
""")
                password_chances -= 1
                
            elif password_chances == 0:
                password_chances -= 1
                print("""
After you tried twice, another mercenary got impatient and opened the door himself.
However, the cargo elevator was fulled with zombies. Unfortunately, he died.
You followed Alice and managed to survive for 120 minutes. 
Congratulations on completing your mission! Your earn 1000 points in this mission！
                """)
            
            else:
                print("\nSomething went wrong.")  
    
    elif 'n' in choice_2_1:
        #Give the path list longer the choice list to make sure it is not easy to get right.
        path = ["right","middle","left","jumping","backstabbing","crawling"]
        
        #Random correct right path for player to guess.
        walkers_path=random.choice(path)

        #Asking the player to guess the walker's path.
        choice_2_2 = input(prompt = f"""
After you kill a zombie, you notice a zombie pouncing on you in front of you, which direction do you want to hide？
(Please input the name of your choice -- the direction.)
Right
Middle
Left
Your choice: """)
        
        #Convert choice_2_2 to lower letters.
        choice_2_2 = choice_2_2.casefold()
        
        #Conditional statement to determine if the player guessed the correct answer
        if choice_2_2 == walkers_path:
            print("""\nYou help the team kill the zombies. You followed Alice and managed to survive for 120 minutes. 
Congratulations on completing your mission! Your earn 1000 points in this mission！""")
            
        elif choice_2_2 != walkers_path and choice_2_2 in path:
            print("\nYou failed to evade the zombie attack. You died.")
            fail()
            
        else:
            print("\nPlease follow the input instruction!!!")
            fail()
    else:
        print("\nPlease follow the input instruction!!!")
        fail()

#Defined the Stage 4. In this stage, Captain still alive. Therefore, the player have a higher possibility to win.
def stage_captain_alive():

    #Long line that indicates the start of a new scene.
    #Warning about error input
    print(f"""
{'*' * 100}
Attention: If you do not follow the input instruction in this stage, it may cause you to fail.
""")
    
    survive_time = 50
    left_time = remaining_time - survive_time
    earn_points = 5000

#Stage Introduction     
    print(f"""
The tech person outside the room who was trying to override the system managed to shut it down before it reach the door.
You successfully saved the Captain Matthew and change the movie plot. You get 5000 points from this action.
What's more, you don't need to stay with a specific target anymore.You still need to survive in this movie for {left_time} minutes.
The team shut down the Red Queen and prepared to leave the Hive.
    """)
    
    #Choose to experience the movie or achieve the survival mission.
    choice_3_1 = input(prompt = f"""
Do you want to keep following the team? <Y/N>
Your choice: """)
    
    #Convert choice to lower letters.
    choice_3_1 = choice_3_1.casefold()
    
    #Conditional statement to choose whether stay with the Captain.
    if 'y' in choice_3_1:
        earn_points += 1000
        print(f"""
Since you saved the captain before, he was very grateful and protected you all the way to help you survive to 120 minutes.
Congratulations! You win the game and earn {earn_points} points!
""")
        
    elif 'n' in choice_3_1:
        print ("\nYou stay at the control room until the survival time countdown is over.")
        
        #Use whlie loop for the time countdown.
        while left_time >= 0:
            if left_time > 25:
                left_time -= 10
                print(f"""
                you still need to survive for {left_time} minutes
                """)
        
            #When left_time = 20, it will tigger a new event.
            elif left_time > 15:
                print("""
A little girl's voice sounded, "Two Lickers are attacking the outer door. It will last forty seconds."
""")
                
                choice_3_2 = input(prompt = f"""
What step you want to take to survive for the rest {left_time} minutes?
(Please input the number or name of your choice.)
1.Stay in the control room.
2.Ask the Red Queen to active the defence system in the corridor.
3.Try to go outside.
Your choice: """)
                
                #Convert choice to lower letters.
                choice_3_2 = choice_3_2.casefold()
                
                #Conditional statement that decides if the player can achieve original novel plot.
                if choice_3_2 == '1' or 'stay' in choice_3_2:
                    print("\nThe Licker break out the door and you died.")
                    fail()
                    break
                
                elif choice_3_2 == '2' or 'red queen' in choice_3_2:
                    earn_points += 1000
                    left_time -= 20
                    
                elif choice_3_2 == '3' or 'outside' in choice_3_2:
                    print("\nYou met the Licker outside and died.")
                    fail()
                    break
                    
                else:
                    print("\nPlease follow the input instruction!!!")
                    fail()
                    break
            else:    
                print(f"""
                                                             ___                  ___            ___                                
                                                            (   )                (   )          (   )      .-.                      
  .--.      .--.    ___ .-.     .--.    ___ .-.      .---.   | |_      ___  ___   | |    .---.   | |_     ( __)   .--.    ___ .-.   
 /    \    /    \  (   )   \   /    \  (   )   \    / .-, \ (   __)   (   )(   )  | |   / .-, \ (   __)   (''")  /    \  (   )   \  
|  .-. ;  |  .-. ;  |  .-. .  ;  ,-. '  | ' .-. ;  (__) ; |  | |       | |  | |   | |  (__) ; |  | |       | |  |  .-. ;  |  .-. .  
|  |(___) | |  | |  | |  | |  | |  | |  |  / (___)   .'`  |  | | ___   | |  | |   | |    .'`  |  | | ___   | |  | |  | |  | |  | |  
|  |      | |  | |  | |  | |  | |  | |  | |         / .'| |  | |(   )  | |  | |   | |   / .'| |  | |(   )  | |  | |  | |  | |  | |  
|  | ___  | |  | |  | |  | |  | |  | |  | |        | /  | |  | | | |   | |  | |   | |  | /  | |  | | | |   | |  | |  | |  | |  | |  
|  '(   ) | '  | |  | |  | |  | '  | |  | |        ; |  ; |  | ' | |   | |  ; '   | |  ; |  ; |  | ' | |   | |  | '  | |  | |  | |  
'  `-' |  '  `-' /  | |  | |  '  `-' |  | |        ' `-'  |  ' `-' ;   ' `-'  /   | |  ' `-'  |  ' `-' ;   | |  '  `-' /  | |  | |  
 `.__,'    `.__.'  (___)(___)  `.__. | (___)       `.__.'_.   `.__.     '.__.'   (___) `.__.'_.   `.__.   (___)  `.__.'  (___)(___) 
                               ( `-' ;                                                                                              
                                `.__.                                                                                               
                
You have managed to survive 120 minutes in the world of Resident Evil 1. 
Congratulations you have earned {earn_points} points!
You achieve the same plot from a Chinese novel called "Terror Infinity".
""")
                break
    else:
        print("\nPlease follow the input instruction!!!")
        fail()

#Defined the Stage 5 as failed stage.
def fail():
    print(f"""
{'*' * 100}
You did not survive in the world of Resident Evil, but thanks for playing!""")
    input(prompt = '<Press any key to exit>\n')    
    
#Calling the first stage in the map to start our game.
start_resident_evil_1()
```
